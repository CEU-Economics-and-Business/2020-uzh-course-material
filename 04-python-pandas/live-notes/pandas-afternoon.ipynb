### Challenge!

use the query syntax to create a function `my_query(dataset, year_start, year_end, state)`

that will return the correct subset of data (not as a no-copy-view of the original data!)

run this for the states 'Nebraska', 'Iowa', 'Vermont', and for the year ranges '2010' to '2015' and '2000' to '2005'

```python
def my_query(dataset, year_start, year_end, state):
    
    out = dataset.query('@year_end >= year >= @year_start & state == @state')\
            .copy()\
            .sort_values(by = ['year', 'period'])
    
    return out
```

```python
df2 = my_query(df, 2010, 2015, 'Iowa')
```
