Repeat this for :

* 04 files are qty unemployed
* 05 files are qty employed
* 06 files are labour force size

hint: write a function that does this (i.e the generator and pd.concat steps) for you

(03 was the unemployment rate)



```python
qty_ue_files = glob.glob(data_path + '*04.csv')
qty_em_files = glob.glob(data_path + '*05.csv')
qty_lf_files = glob.glob(data_path + '*06.csv')
```

```python
def append_data(file_list, col_list):
    
    data_gen = ( pd.read_csv(iFile,
                             usecols = col_list,
                             index_col = ['state', 'year', 'period']
                            )
                for iFile in file_list
                )
    return pd.concat(data_gen)
```

```python
qty_ue = append_data(qty_ue_files, col_list = cols)
qty_em = append_data(qty_em_files, col_list = cols)
qty_lf = append_data(qty_lf_files, col_list = cols)
rte_ue = append_data(ue_files, col_list = cols)
```

```python
qty_ue.head()
```

```python
rte_ue.head()
```

```python
## renaming cols

rte_ue.rename(columns = {'value': 'ue_rate'},
              inplace = True
             )
qty_ue.rename(columns = {'value': 'ue_qty'},
              inplace = True
             )
qty_lf.rename(columns = {'value': 'lf_qty'},
              inplace = True
             )
qty_em.rename(columns = {'value': 'em_qty'},
              inplace = True
             )
```

```python
merged_data = rte_ue.merge(qty_ue,
                           left_index = True,
                           right_index= True,
                           how = 'inner'
                            )
```

```python
from functools import reduce
```
